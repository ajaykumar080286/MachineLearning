```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X,y =make_classification(n_samples=5, n_features=2, n_classes=2,n_informative=2, n_redundant=0, random_state=0)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
plot_tree(clf)
```

```python
clf.feature_importances_
```
