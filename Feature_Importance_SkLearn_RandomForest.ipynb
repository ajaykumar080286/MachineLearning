```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('mnist_train.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
sns.heatmap(rf.feature_importances_.reshape(28,28))
```
