```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import plotly.express as px
```

```python
df=pd.read_csv("student_clustering.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['iq'])
```

```python
wcss=[]
```

```python
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit_predict(df)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
x=df.iloc[:,:].values
```

```python
km=KMeans(n_clusters=4)
y_means=km.fit_predict(x)
```

```python
x[y_means==3,1]
```

```python
plt.scatter(x[y_means==0,0], x[y_means==0,1], color='blue')
plt.scatter(x[y_means==1,0], x[y_means==1,1], color='red')
plt.scatter(x[y_means==2,0], x[y_means==2,1], color='green')
plt.scatter(x[y_means==3,0], x[y_means==3,1], color='yellow')
```

**KMeans for 3 D**

```python
from sklearn.datasets import make_blobs
```

```python
centriod=[(-5,-5,-5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std=[1,1,1,1]
```

```python
X,y=make_blobs(n_samples=200, cluster_std=cluster_std, centers=centriod, n_features=3, random_state=1)
```

```python
X
```

```python
import plotly.express as px
fig=px.scatter_3d(x=X[:,0],y=X[:,1], z=X[:,2])
fig.show()
```

```python
wcss=[]
for i in range(1,21):
  km=KMeans(n_clusters=i)
  km.fit_predict(X)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
kd=KMeans(n_clusters=4)
y_pred1=kd.fit_predict(X)
```

```python
df1=pd.DataFrame(X)
```

```python
df1['col1']=X[:,0]
df1['col2']=X[:,1]
df1['col3']=X[:,2]
df1['lable']=y_pred1
```

```python
df1
```

```python
fig=px.scatter_3d(df1, x='col1', y='col2', z='col3', color='lable')
fig.show()
```
