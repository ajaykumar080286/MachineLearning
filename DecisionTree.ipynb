```python
def analysize(max_depth):
  df=pd.read_csv("Social_Network_Ad.csv")

  X=df.iloc[:,2:4].values
  y=df.iloc[:,-1].values

  clf=DecisionTreeClassifier(max_depth=max_depth)
  clf.fit(X,y)

  a=np.arange(start=X[:,0].min()-1, stop= X[:,0].max()+1,step=0.1)
  b=np.arange(start=X[:,1].min()-1, stop= X[:,1].max()+1,step=100)

  XX,YY=np.meshgrid(a,b)

  input_array=np.array([XX.ravel(), YY.ravel()]).T

  lables=clf.predict(input_array)

  plt.contourf(XX,YY,lables.reshape(XX.shape), alpha=0.5)
  plt.scatter(X[:,0],X[:,1],c=y)





```

```python
analysize(None)
```

```python
analysize(1)
```

```python
analysize(2)
```

```python
analysize(4)
```
