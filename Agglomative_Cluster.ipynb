```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
df= pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data (1).csv")
```

```python
df.head()
```

```python
data=df.iloc[:,3:5]
```

```python
data.head()
```

```python
plt.figure(figsize=(10,7))
plt.title("Customer Dendogram")

dnd=sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
cluster=AgglomerativeClustering(n_clusters=5,metric='euclidean', linkage='ward')
```

```python
label_=cluster.fit_predict(data)
```

```python
label_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=cluster.labels_,cmap='rainbow')
```
