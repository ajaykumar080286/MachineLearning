```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic (1).csv")
```

```python
df.head(5)
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title("passenger traveller with")
```

```python
#extract numerical part
df['number_numercial']=pd.to_numeric(df['number'], errors='coerce',downcast='integer')
```

```python
df['number_categorical']= np.where(df['number_numercial'].isnull(), df['number'], np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df.head()
```

```python
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+\.?\d*)')
```

```python
df['cabin_cat'] = df['Cabin'].str.extract(r'([a-zA-Z]+)')
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot.bar()
```

```python
df.head()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])
```
