```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(df.drop("has_covid", axis=1), df["has_covid"],test_size=0.2,random_state=0)
```

```python
#Simple Imputer for fever column to fill missing values
```

```python
sm=SimpleImputer()
```

```python
sm
```

```python
X_train_fever=sm.fit_transform(X_train[['fever']])
X_test_fever=sm.fit_transform(X_test[['fever']])
```

```python
X_train_fever.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
X_train_new=transformer.fit_transform(X_train).shape
X_test_new=transformer.fit_transform(X_test).shape
```
