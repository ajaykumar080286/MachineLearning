```python
from numpy.ma import shape
class ownBatchRegressor:

  def __init__(self,lr,epochs,bs):
    self.intercept_=None
    self.coef_=None
    self.lr=lr
    self.epochs=epochs
    self.bs=bs

  def fit(self,X_train,y_train):

    # calculation of intercept

    self.intercept_=0
    self.coef_=np.ones(X_train.shape[1])
    batc_size=int(X_train.shape[0]/self.bs)
    print(batc_size)

    for i in range(self.epochs):
      for j in range(int(X_train.shape[0]/self.bs)):


            idx = random.sample(range(X_train.shape[0]), batc_size)

            y_hat=np.dot(X_train[idx],self.coef_)+self.intercept_
            der_slope=-2*np.mean(y_train[idx]-y_hat)
            self.intercept_=self.intercept_ -(self.lr*der_slope)
            # coef
            der_coef_slope=-2*np.dot((y_train[idx]-y_hat),X_train[idx])
            self.coef_=self.coef_- self.lr*der_coef_slope

    print(self.intercept_)
    print(self.coef_)

  def predict(self,X_test):
    return np.dot(X_test,self.coef_)+self.intercept_
```
