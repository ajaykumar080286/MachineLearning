```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("placement (1).csv")
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.kdeplot(df['cgpa'])

plt.subplot(1,2,2)
sns.kdeplot(df['placement_exam_marks'])

plt.show()
```

```python
print("mean of Cgpa", df["cgpa"].mean())
print("Standard Deviation of Cgpa", df["cgpa"].std())
print("Min value of Cgpa", df["cgpa"].min())
print("Max value of Cgpa", df["cgpa"].max())
```

```python
# find the boundayr value
lower_boundry=df['cgpa'].mean()-3*df['cgpa'].std()
upper_boundry=df['cgpa'].mean()+3*df['cgpa'].std()
```

```python
print("Lower allowed",lower_boundry)
print("Highest allowed",upper_boundry)
```

```python
#find the outlier

df[(df['cgpa']>8.80)   | (df['cgpa'] < 5.11)]
```

```python
#Trimming
new_df=df[(df['cgpa']<8.80)   & (df['cgpa'] >5.11)]

new_df
```

```python
df['z_score']=(df['cgpa']-df['cgpa'].mean())/df['cgpa'].std()
```

```python
df
```

```python
df[(df['z_score']>3)   | (df['z_score'] < -3)]
```

```python
#Trimming
new_df=df[(df['z_score']<3)   & (df['z_score'] >-3)]

new_df
```

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
df['cgpa']=np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
      df['cgpa']<lower_limit,
      lower_limit,df['cgpa']
      )
    )
```
