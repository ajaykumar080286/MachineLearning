```python
df=pd.read_csv('student_clustering.csv')
X=df.iloc[:,:].values
```

```python
from types import new_class
class Kmeans:

  def __init__(self, n_cluster=2, max_iter=100):
    self.n_cluster=n_cluster
    self.max_iter=max_iter
    self.centriods=None

  def fit_predict(self,X):
    # Select Random Centriod
    random_index=random.sample(range(0,X.shape[0]),self.n_cluster)
    self.centriods=X[random_index]

    # Assign clister
    for i in range(self.max_iter):
      cluster_group=self.assign_cluster(X)
      #move centriod
      old_centriod=self.centriods
      self.centriods=self.move_centriod(X,cluster_group)
      #finsih
      if(old_centriod==self.centriods).all():
        break

    return cluster_group


  #Move centriod and calculate new centriod
  def move_centriod(self,X,cluster_group):
    new_centriod=[]
    cluster_type=np.unique(cluster_group)
    for type in cluster_type:
      new_centriod.append(X[cluster_group==type].mean(axis=0))

    return np.array(new_centriod)


  #Assign Cluster
  def assign_cluster(self,X):

    cluster_group=[]
    distances=[]
    sum_dis=[]

    for row in X:
      for centriod in self.centriods:
        distances.append(np.sqrt(np.dot(row-centriod, row-centriod)))
      min_distance=min(distances)
      index_pos=distances.index(min_distance)
      cluster_group.append(index_pos)
      distances.clear()

    return np.array(cluster_group)
```

```python
km=Kmeans(n_cluster=4, max_iter=300)
```

```python
y_means=km.fit_predict(X)
```

```python
plt.scatter(X[y_means==0,0], X[y_means==0,1], color='red')
plt.scatter(X[y_means==1,0], X[y_means==1,1], color='blue')
plt.scatter(X[y_means==2,0], X[y_means==2,1], color='yellow')
plt.scatter(X[y_means==3,0], X[y_means==3,1], color='green')


plt.show()
```
